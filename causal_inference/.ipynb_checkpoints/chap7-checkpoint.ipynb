# 重回帰モデルにおける6つの仮定
1. $E[\varepsilon] = 0$  誤差項の期待値はゼロ
2. 母数の線形性
3. $E[\varepsilon|X] = 0$   誤差項の条件付き期待値はゼロ
4. 完全な多重共線性がない
5. 誤差項の分散均質性の仮定
6. 誤差項の正規性の仮定

## 仮定1. $E[\varepsilon] = 0$ 誤差項の期待値ゼロ
単回帰モデル  $y=\alpha_0 + \beta_1X_1+u$  で  
誤差項uの期待値  $E[u]=\delta, \delta\neq0$  の場合を考える.

単回帰モデルの式を変形していく.  
$\quad y=\alpha_0 + \beta_1X_1+u$  
$\quad y=\alpha_0 + \beta_1X_1+u + \delta - \delta$  
$\quad y=(\alpha_0 + \delta) + \beta_1X_1+ (u - \delta)$  
ここで$(\alpha_0 + \delta)=\beta_0, (u - \delta)=\varepsilon$とおくと、  
$\quad y=\beta_0 + \beta_1X_1+ \varepsilon$  

新たに定義した誤差項の期待値は  
$\quad E[\varepsilon]= E[u-\delta] = E[u] - E[\delta]$  
定義$E[u]=\delta$であり、また$E[k] = k (k:定数)$より  
$\quad E[u] - E[\delta] = \delta - \delta = 0$


これを実際にRで試してみる

## 参考文献
- [Rのlm関数とは](https://ai-trend.jp/programming/r-beginner/r-3/)

````r
set.seed(1);  n1<- 1000 #シード値固定、サンプルサイズn1=1000
a0 <-1; b1 <- 1.5 #回帰係数(切片、傾き)を設定

x1 <- rnorm(n1) # ε = x1~N(0,1)  標準正規分布にもとづく乱数x1を生成
u1 <- rnorm(n1, 10, 1)  # u1 ~N(10, 1)期待値!=0の誤差項を作る

y1 <- a0 + b1*x1 + u1 #単回帰モデル
model1 <- lm(y1~x1) #lm(linear model)関数で係数を算出 model.fit

summary(model1) #結果表示
````

CoefficientsのEstimate列 Intercept行をみると、a0が1から10に(u1の期待値分)変化した.  
またb1(x1のestimate)は1.5のままで誤差項の期待値の影響を受けていないことがわかる

## 仮定2 母数における線形性
$\quad X_1 \sim N(0,1)$  
$\quad X_2 \sim LN(0,1)$  ($X_2$は対数正規分布に従う)  
の2変数があり、  
$\quad \varepsilon \sim N(0,1)$  
$\quad \beta_0 (=  1.0), \quad \beta_1 (=  1.5)$としたとき、  

以下の4式は全て$\gamma$にあたえるXの平均処置効果が$\beta_1=1.5$であることを意味している。  
$\quad \gamma_1 = \beta_0 + \beta_1X_1+\varepsilon \quad\quad\quad\quad\quad\quad (7.1)$   
$\quad \gamma_2 = \beta_0 + \beta_1log(X_2)+\varepsilon \quad\sim N()\quad (7.2)$   
$\quad \gamma_3 = exp(\beta_0 + \beta_1X_1+\varepsilon) \quad\sim LN() ~(7.3)$   
$\quad \gamma_4 = \beta_0X_2^{\beta_1}e^\varepsilon\quad\quad\quad\quad\quad\quad\quad\quad~~ (7.4)$   

````r
#実際にB1=1.5になるかやってみる
rm(list = ls()) #ワークスメモリをクリア
data07a <- read.csv('./exc_data//data07a.csv') # csv読み込み
summary(data07a) #X1, X２と、式(7,1)~(7.4)に従う結果y1~y4が入っている
````

### 線形モデルで式(7.2~4)を表す形で再チャレンジ

### 対数変換後の回帰係数の解釈

$\quad (7.2)~\gamma_2 = \beta_0 + \beta_1log(X_2)+\varepsilon$   

$\quad (7.3)~\gamma_3 = exp(\beta_0 + \beta_1X_1+\varepsilon)$   
$\quad\quad \leftrightarrow log(\gamma_3) = \beta_0 + \beta_1X_1+\varepsilon$  

$\quad (7.4)~\gamma_4 = \beta_0X_2^{\beta_1}e^\varepsilon$   
$\quad\quad \leftrightarrow log(\gamma_4) = log(\beta_0X_2^{\beta_1}e^\varepsilon) = log(\beta_0) + \beta_1log(X_2)  + \varepsilon log(e)$  
$\quad\quad \leftrightarrow log(\gamma_4)  = log(\beta_0) + \beta_1log(X_2)  + \varepsilon$  
 

A: $y = \beta_0 + \beta_1x$  を$x$で微分して    
$\quad \frac{d}{dx}y = \beta_1\quad$   ```xが1単位変化した時yはB1単位変化する```

B: $y = \beta_0 + \beta_1log(x)$  を$x$で微分して  
公式$\frac{d}{dx}logx = \frac{1}{x}$より   
$\quad \frac{d}{dx}y = \beta_1\frac{1}{x}\quad$    ```xが1%変化した時yはB1/100単位変化する (xの変化率が0.01=1%のとき、yは0.0B1=B1%変化)```

C: $log(y) = \beta_0 + \beta_1x$  を$x$で微分して  
$\quad \frac{d}{dx}log(y) = \beta_1$  
$\quad \beta_1 = \frac{d}{dy}log(y)\frac{dy}{dx} = \frac{1}{dy}\frac{dy}{dx}$  
$\quad \frac{dy}{y} = \beta_1dx\quad$  ```xが1単位変化した時yは100*B1%変化する (yの変化率はB1=100*B1%)```

D: $log(y) = \beta_0 + \beta_1log(x)$  を$x$で微分して  
$\quad \frac{d}{dx}log(y) = \beta_1\frac{1}{x}$  
$\quad \frac{\beta_1}{x} = \frac{d}{dy}log(y)\frac{dy}{dx} = \frac{1}{y}\frac{dy}{dx}$   
$\quad \frac{dy}{y} = \beta_1\frac{dx}{x}\quad$  ```xが1%変化した時yはB1%変化する```   


````r
#教科書のテイラー展開でもやってみる
# xが十分小さいときのテイラー展開の近似を確認
x1 <- 0.1
log(1 + x1)
x1- 1/2* x1^2 + 1/3 * x1^3 - 1/4 * x1^4 # ~=log(1+x) xが十分小さい時log(1+x)~=x
x1
````

  
### 変数の線形性と母数の線形性の違い
母数の線形性は、変数を変換して線形にできていれば保たれる。そのため変数が(変換すれば線形になるのに)非線形な状態でも問題ない  
OK: $\quad \gamma = \beta_0X_1^{\beta_1}X_2^{\beta_2}e^{\varepsilon}\quad$  `対数変換すれば線形`   
NG: $\quad \gamma = \beta_0X_1^{\beta_1}X_2^{\beta_2}+\varepsilon\quad$     `どうやっても線形にならない`

````r
set.seed(1);  n1<- 1000 #シード値固定、サンプルサイズn1=1000
b0 <- 1; b1 <- 0.6; b2 <- 0.4 #回帰係数(切片、傾き)を設定

x3 <- runif(n1); x4 <- runif(n1); e1 <- rnorm(n1) # x3, x4~一様分布, e ~N(0,1)で乱数を生成

y5 <- b0 * x3^b1 * x4^b2 * exp(e1) # OKの式
y6 <- b0 * x3^b1 * x4^b2 + e1 # NGの式

model9 <- lm(log(y5)~log(x3) + log(x4)) # OKの式でlogとった式: log(y5) = log(全部かけ)
model11 <- lm(log(y6)~log(x3) + log(x4)) # NGを変換してもたどり着かない式 log(y6) = log(BXX + e) != log(BXX)+log(e)
````

````r
summary(model9) #b1=0.6, b2=0.4にまあまあ近い
````

````r
summary(model11) #b1=0.6, b2=0.4にならない+エラー出る(真数<0のため。 真数はloga(b)のb)
````

### 多変量における診断方法
単回帰であればxyのプロットすれば合うモデルを選択できるが、多変量になるとそうもいかない。そこで**成分プラス残差プロット**(偏残差プロット)がおすすめ  
３変量で考える  
$\quad y_{1i} = exp[\beta_0 + \beta_1x_{1i} + \beta_2log(x_{3i})+\varepsilon_i]\quad (7.12)$ 

$\quad\quad \beta_0=1.0, \beta_1=0.5, \beta_2=0.2$   
$\quad\quad x_1, x_2\sim N(5, 1), r_{1,2}=0.5$   
$\quad\quad x_3 = exp(x_2)$   
$\quad\quad \varepsilon \sim N(0, 0.01)$   

j番目の説明変数$x_j$の偏残差を以下で定義する.   
$\quad e_{ij} = \beta_jx_{ji} + e_i \quad (7.13)$ 

````r
#データ読み込み
rm(list = ls())
data07b <- read.csv('./exc_data//data07b.csv')
summary(data07b) #x3=exp(X２)
````

## 仮定3 $E[\varepsilon|X] = 0$   誤差項の条件付き期待値はゼロ
**だいじ** 最小二乗法で回帰係数を推定するための前提になる仮定  

誤差項$\varepsilon$と共変量Xは独立と仮定すると、
$\quad E[\varepsilon|X] = E[\varepsilon]\quad$(誤差項は共変量と**平均独立**)  
仮定1より$E[\varepsilon|] = 0$より
$\quad E[\varepsilon|X] = 0\quad$  

これは目的変数に影響を与える交絡因子が十分にモデルに含まれていなければならない、という前提と同じこと(ただし診断する術はない) *データのクレンジング・異常検知がんばろうね*  
また次の式を無視可能な割り付けになっている前提ともいえる
$\quad Pr(T|\gamma(1), \gamma(0), X) = Pr(T|X)$

### 不要な変数をモデルに取り入れる問題
結論入れてもOK (ただし不要な変数が他の使える変数と相関が強いと標準誤差が大きくなるおそれがある =多重共線性)  

つぎの例でやってみる   
$\quad \gamma = \beta_0 + \beta_1X_1 + \beta_2 X_2 + \beta_3X_3 + \varepsilon$  
$\quad\quad \beta_3=0  ~(X_3$は不要な変数)  
$\quad\quad \beta_0=1,\beta_1=1.3,\beta_2=1.2$    

````r
# やってみよー
rm(list = ls())
data07c <- read.csv('./exc_data//data07c.csv')
summary(data07c)
````

````r
model17 <- lm(y1 ~ x1, data=data07c) #y = B0 + B1X1 + e
model18 <- lm(y1 ~ x1 + x2, data=data07c) #y = B0 + B1X1 + B2X2 + e
model19 <- lm(y1 ~ x1 + x2 + x3, data=data07c) #y = B0 + B1X1 + B2X2 + B3X3 + e
````

````r
summary(model17) #B1=1.3からずれる
````

````r
summary(model18) #B1=1.3, B2=1.2になる
````

````r
summary(model19) #B1=1.3, B2=1.2になる. 一方X2の標準誤差(Std.Error)がmodel18より大きくなっている
````

````r
# X1,2とX1,2,3をつかった重回帰モデルの信頼区間をだす
confint(model18)
confint(model19)
````

モデル18, 19のx1の行を見ると信頼区間が1.2~1.4でどちらも$\beta_1=1.3$を含んでいるので正しい.  
しかしモデル19の方が信頼区間の幅が広くなっていて、標準誤差が大きくなっている = X3のノイズで推定精度が悪化している.  
推定結果が上下にずれることはない(不偏)が、ノイズは入ることを承知の上で、迷ったら入れろ  

#### 情報量基準による変数選択は因果推論では推奨されない  
表7.18を見るとモデルAICではカバー率(モンテカルロ法で1万回シミュレーションしたときに信頼区間に真値が何%含まれていたか)が真のモデルより小さい。  
いっぽうモデル19では真のモデルと同等のカバー率を維持できている　。(標準誤差は上がるが)

### 中間変数をモデルに入れる問題
**中間変数(因果関係の間に位置する変数)をモデルに含めてはならない**

$\quad \quad X_2$  
$\quad ↗️  \quad↘️$  
$X1\quad →\quad ~\gamma$  

X1<-X2&Y<-X2ではなく、X1->X2->Yになっている状態.  
X1->Yを**直接効果**, X1->X2->Yを**間接効果**といい、  
全体の効果 =  直接効果 + 間接効果  
すなわち  
$\quad \quad X_2 =\alpha_0 + \alpha_1X_1 + \varepsilon_1$  
$\quad \quad \gamma = \beta_0 + \beta_1X_1 + \beta_2 X_2 + \varepsilon_2$  

$\quad \leftrightarrow \gamma = \beta_0 + \beta_1X_1 + \beta_2 (\alpha_0 + \alpha_1X_1 + \varepsilon_1) + \varepsilon_2$  
$\quad \quad \quad = (\beta_0 +\beta_2\alpha_0) + (\beta_1 + \beta_2\alpha_1)X_1 + (\beta_2\varepsilon_1 + \varepsilon_2)$ 




````r
#中間変数の何が問題か実際に見てみよう
rm(list = ls())
data07d <- read.csv('./exc_data//data07d.csv')
summary(data07d)
````

````r
model20 <- lm(y1 ~ x1, data=data07d) #y = B0 + B1X1 + e
model21 <- lm(y1 ~ x1 + x2, data=data07d) #y = B0 + B1X1 + B2X2 + e, X2はX1->Yの中間変数
````

````r
summary(model20)
````

````r
summary(model21)
````

モデル20では正しく推定できていたものの、モデル21ではできなくなっている
